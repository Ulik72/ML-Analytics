# target 0

```python
data_target_0=data.iloc[:,:-1]
data_target_0


def correct(rows):
    if rows=='gas1':
        return 1
    else:
        return 2

data_target_0['feature4'] = data_target_0['feature4'].apply(correct)
data_target_0
```

```python
# Кластеризируем Target0 на два класса
from sklearn.preprocessing import StandardScaler
x=data_target_0.iloc[:,:-1]
x=np.nan_to_num(x)
scaler= StandardScaler().fit_transform(x) 
scaler
```

```python
from sklearn.cluster import KMeans
clasternum=2
kmeans = KMeans(init="random", n_clusters=clasternum, n_init=10,max_iter=300, random_state=42)
kmeans.fit(scaler)
label=kmeans.labels_
```

```python
data_target_0['Claster']=label
```

# Кластер 0

```python
data_target_0_claster_0=data_target_0[data_target_0['Claster']==0]
```

```python
plt.figure(figsize=(20,8))

plt.subplot(1,2,1)
plt.title('target0')
sns.distplot(data_target_0_claster_0.target0)

plt.subplot(1,2,2)
plt.title('R and median')
sns.boxplot(y=data_target_0_claster_0.target0)

plt.show()
```

```python
# Удалим выбросы
data_target_0_claster_0=data_target_0_claster_0[data_target_0_claster_0['target0']>11]
data_target_0_claster_0

plt.title('Boxplot')
sns.boxplot(y=data_target_0_claster_0.target0)

plt.show()
```

# Обучим модель под этот кластер

```python
from catboost import CatBoostRegressor
from catboost import Pool
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
# Данные
x=data_target_0_claster_0.iloc[:,:-2] # Зависимые переменные 
y=data_target_0_claster_0['target0']# Прогноз

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20)

# Создаем Pool для передачи в CatBoost
train_pool=Pool(data=x_train, label=y_train)
validation_pool=Pool(data=x_test,label=y_test)
```

```python
# Объявляем и обучаем модель / высчитываем дополнительные метрики и строим графики обучения
model_Cat_1=CatBoostRegressor(
    iterations=3000,
    learning_rate=0.003,
    custom_metric=['R2','MAE']
)

model_Cat_1.fit(
    train_pool, 
    eval_set=validation_pool,
    verbose=False,
    plot=True
);
```

```python
pred=model_Cat_1.predict(x_test)
print(round(metrics.mean_squared_error(y_test,pred),2),"MSE")
print(round(metrics.median_absolute_error(y_test,pred),2),"MAE")
print(round(metrics.r2_score(y_test,pred),2),"R2")
print(round(metrics.mean_absolute_percentage_error(y_test,pred),2),"MAPE")
```

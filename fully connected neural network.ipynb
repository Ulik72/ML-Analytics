```python
import torch
import numpy as np
import pandas as pd
```

```python
data=pd.read_excel(r'C:\Users\Julaysa\Машинное обучение\Задачи регрессии\Fakticheskie_dannye.xlsx')
#Нормализация данных
from sklearn import preprocessing
data_x=data.iloc[:,:14] #Нормализуем x
data_x = preprocessing.normalize(data_x)
data_x = pd.DataFrame(data_x)
```

```python
# Данные
x=data_x.iloc[:,:14]
y=data.iloc[:,-1]



from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20)



# конвертируем данные
x_train=torch.tensor(x_train.values).type(torch.FloatTensor)
y_train=torch.tensor(y_train.values).type(torch.FloatTensor)
x_test=torch.tensor(x_test.values).type(torch.FloatTensor)
y_test=torch.tensor(y_test.values).type(torch.FloatTensor)


#Транспонируем векторы-строки в векторы-столбцы
x_train.unsqueeze_(1)
y_train.unsqueeze_(1);
```

```python
class SineNet(torch.nn.Module):
    def __init__(self, n_hidden_neurons):
        super(SineNet, self).__init__()
        self.fc1 = torch.nn.Linear(14, n_hidden_neurons) #первый слой 
        self.act1 = torch.nn.Sigmoid() # функция активации 
        self.fc2 = torch.nn.Linear(n_hidden_neurons, 1) #второй слой 

    def forward(self, x):
        x = self.fc1(x)
        x = self.act1(x)
        x = self.fc2(x)
        return x

sine_net = SineNet(1500) # Колличество нейронов 
```

```python
def predict(net, x, y):
    y_pred = net.forward(x)
```

```python
# Оптимизатор градиентного спуска  
optimizer = torch.optim.Adam(sine_net.parameters(), lr=0.001)
```

```python
# Функция потерь
loss=torch.nn.MSELoss()
```

```python
for epoch_index in range(1000):
    optimizer.zero_grad() # Обнуляем градиенты
    y_pred = sine_net.forward(x_train)
    loss_val = loss(y_pred, y_train) # Функция потерь
    loss_val.backward() # Производная
    optimizer.step()
    print("Эпоха",epoch_index,"Функция потерь",loss_val)
```
